```python
import math
import numpy as np
import matplotlib.pyplot as plt

RADIUS = 1.0
DISTANCE = 1.0

theta = np.linspace(0, 2 * math.pi, 1000)

# Cirklar
x1 = RADIUS * np.cos(theta)
y1 = RADIUS * np.sin(theta)

x2 = DISTANCE + RADIUS * np.cos(theta)
y2 = RADIUS * np.sin(theta)

# Centrum
c1 = (0.0, 0.0)
c2 = (DISTANCE, 0.0)
cm = (DISTANCE / 2.0, 0.0)

plt.figure(figsize=(6, 6))
plt.plot(x1, y1)
plt.plot(x2, y2)

plt.scatter([c1[0], c2[0], cm[0]], [c1[1], c2[1], cm[1]], zorder=3)

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Bild 2 — Centrum-noder")
plt.axis("off")
plt.show()
```

```python
# Intersektionspunkter för två cirklar med samma radie och avstånd = radie
h = math.sqrt(3) / 2 * RADIUS
ip1 = (DISTANCE / 2.0,  h)
ip2 = (DISTANCE / 2.0, -h)

plt.figure(figsize=(6, 6))
plt.plot(x1, y1, alpha=0.5)
plt.plot(x2, y2, alpha=0.5)

# Centrum
plt.scatter([c1[0], c2[0]], [c1[1], c2[1]], zorder=3)

# Intersektioner
plt.scatter([ip1[0], ip2[0]], [ip1[1], ip2[1]], zorder=4)

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Bild 3 — Intersektionspunkter")
plt.axis("off")
plt.show()
```

```python
# Irrationell höjd i Vesica Piscis
h = math.sqrt(3) / 2 * RADIUS

plt.figure(figsize=(6, 6))
plt.plot(x1, y1, alpha=0.3)
plt.plot(x2, y2, alpha=0.3)

# Vertikal linje som visar irrationell höjd
plt.plot(
    [DISTANCE / 2.0, DISTANCE / 2.0],
    [-h, h],
    linestyle="--"
)

# Centrum
plt.scatter([c1[0], c2[0]], [c1[1], c2[1]], zorder=3)

# Intersektioner
plt.scatter([ip1[0], ip2[0]], [ip1[1], ip2[1]], zorder=4)

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Bild 4 — Irrationell funktion (√3 / 2)")
plt.axis("off")
plt.show()
```

```python
scales = [0.5, 1.0, 2.0, 5.0]
ratios = []

for s in scales:
    h_s = math.sqrt(3) / 2 * s
    ratios.append(h_s / s)

plt.figure(figsize=(6, 4))
plt.plot(scales, ratios, marker="o")
plt.axhline(math.sqrt(3) / 2, linestyle="--")

plt.xlabel("Skala")
plt.ylabel("Höjd / Radie")
plt.title("Bild 5 — Irrationell funktion är skaloberoende")
plt.grid(True)
plt.show()
```

```python
# Två raka sträckor med samma längd (1.0)
straight_a = RADIUS          # radie
straight_b = DISTANCE        # centrumavstånd

# Den relationellt uppkomna höjden
h = math.sqrt(3) / 2 * RADIUS

# Absolut och relativ skillnad
absolute_diff = abs(straight_a - h)
relative_diff = absolute_diff / straight_a

{
    "rak_stracka_A": straight_a,
    "rak_stracka_B": straight_b,
    "relationell_hojd": h,
    "absolut_skillnad": absolute_diff,
    "relativ_skillnad": relative_diff
}
```

```python
plt.figure(figsize=(6, 6))

# Baslinje (rak sträcka)
plt.plot([0, 1], [0, 0], linewidth=3)

# Relationell höjd
plt.plot([0.5, 0.5], [0, h], linewidth=3)

# Markeringar
plt.scatter([0, 1], [0, 0], zorder=3)
plt.scatter([0.5], [h], zorder=4)

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Bild 6–7 — Samma längd, olika resultat")
plt.axis("off")
plt.show()
```

```python
# Fyra primära riktningar/funktioner runt centralpunkten
# (utan operatorer, utan dynamik)

cx, cy = cm  # centralpunkt (mitt mellan centrum)

directions = {
    "in":    (cx, cy - h),
    "out":   (cx, cy + h),
    "left":  (cx - RADIUS/2, cy),
    "right": (cx + RADIUS/2, cy)
}

plt.figure(figsize=(6, 6))
plt.plot(x1, y1, alpha=0.3)
plt.plot(x2, y2, alpha=0.3)

# Centralpunkt
plt.scatter([cx], [cy], zorder=5)

# Fyra grundfunktioner (geometriska riktningar)
for p in directions.values():
    plt.plot([cx, p[0]], [cy, p[1]], linewidth=2)

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Bild 8 — Fyra grundfunktioner (centralisering)")
plt.axis("off")
plt.show()
```

```python
# Sju stabila noder i Nivå I
nodes_x = [
    c1[0],      # vänster centrum
    c2[0],      # höger centrum
    ip1[0],     # övre intersektion
    ip2[0],     # nedre intersektion
    cm[0],      # centralpunkt
    c1[0] - RADIUS,  # vänster yttergräns
    c2[0] + RADIUS   # höger yttergräns
]

nodes_y = [
    c1[1],
    c2[1],
    ip1[1],
    ip2[1],
    cm[1],
    0.0,
    0.0
]

plt.figure(figsize=(6, 6))
plt.plot(x1, y1, alpha=0.3)
plt.plot(x2, y2, alpha=0.3)

plt.scatter(nodes_x, nodes_y, zorder=5)

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Bild 9 — Sju tillåtna noder (Nivå I)")
plt.axis("off")
plt.show()
```

```python
# Absolut och procentuell relation mellan rak sträcka och relationell höjd

rak_stracka = RADIUS                 # referenslängd = 1.0
relationell_hojd = math.sqrt(3) / 2 * RADIUS

absolut_relation = relationell_hojd - rak_stracka
procent_relation = (absolut_relation / rak_stracka) * 100.0

{
    "rak_stracka": rak_stracka,
    "relationell_hojd": relationell_hojd,
    "absolut_relation": absolut_relation,
    "procentuell_relation": procent_relation
}
```

```python
plt.figure(figsize=(6, 4))

labels = ["Rak sträcka (100%)", "Relationell höjd"]
values = [rak_stracka, relationell_hojd]

plt.bar(labels, values)
plt.axhline(rak_stracka, linestyle="--")

plt.title("Procentuell relation: rak sträcka vs relationell höjd")
plt.ylabel("Längd")
plt.show()
```

```python
# Försök att lägga till en åttonde nod inom samma klass (Nivå I)
# Vi placerar den symmetriskt på vertikalaxeln bortom intersektionspunkterna

attempt_8 = (cm[0], 2 * h)

plt.figure(figsize=(6, 6))
plt.plot(x1, y1, alpha=0.3)
plt.plot(x2, y2, alpha=0.3)

# Befintliga sju noder
plt.scatter(nodes_x, nodes_y, zorder=5)

# Försökt åttonde nod
plt.scatter([attempt_8[0]], [attempt_8[1]], marker="x", s=120, zorder=6)

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Bild 10 — Försök till åttonde nod (klassbrott)")
plt.axis("off")
plt.show()
```
